```python
import pandas as pd

newdf = pd.read_csv('D:\\Master of DS\\customer.csv')

# Create a list to store the rows
rows = []

for customer_id in newdf['cust_id'].unique():
    # Select transactions for the current customer
    customer_transactions = newdf[newdf['cust_id'] == customer_id]
    
    # Calculate total transaction amount
    total_transaction = customer_transactions['total'].sum()
    
    # Find preferred category (assuming it's a categorical column)
    preferred_category = customer_transactions['category'].mode().iloc[0]
    
    # Find preferred payment method (assuming it's a categorical column)
    preferred_payment_method = customer_transactions['payment_method'].mode().iloc[0]
    
    # Get the latest order date
    latest_order_date = customer_transactions['order_date'].max()
    
    # Retain age and gender
    age = customer_transactions['age'].iloc[0]  # Assuming age is constant for a customer
    gender = customer_transactions['Gender'].iloc[0]  # Assuming gender is constant for a customer
    location = customer_transactions['State'].iloc[0] 
    full_name = customer_transactions['full_name'].iloc[0]
     # Calculate the total number of transactions
    total_purchase = len(customer_transactions)
    
    # Create a new row for the customer in the list
    rows.append({
        'customer_id': customer_id,
        'total_transaction': total_transaction,
        'preferred_category': preferred_category,
        'preferred_payment_method': preferred_payment_method,
        'latest_order_date': latest_order_date,
        'age': age,
        'gender': gender,
        'location': location,
        'full_name': full_name,
        'TotalPurchases': total_purchase
    })
    
# Create a new DataFrame from the list of rows
new_df = pd.DataFrame(rows)

new_df
```

```python
new_df.to_csv('D:\\Master of DS\\customer.csv', index=False)
```

```python
import numpy as np

# Calculate quartiles for total_transaction
first_quartile = new_df['total_transaction'].quantile(0.25)
second_quartile = new_df['total_transaction'].quantile(0.5)
third_quartile = new_df['total_transaction'].quantile(0.75)

# create a list of our conditions
conditions = [
    (new_df['total_transaction'] > 10000),
    (new_df['total_transaction'] > third_quartile),
    (new_df['total_transaction'] > second_quartile),
    (new_df['total_transaction'] > first_quartile),
    (new_df['total_transaction'] <= first_quartile)
]

values = ['platinum', 'gold', 'silver', 'bronze', 'standard']

new_df['membership_level'] = np.select(conditions, values)
```

```python
np.random.seed(42)
missing_percentage = 2.7
num_rows = new_df.shape[0]
num_missing = int((missing_percentage / 100) * num_rows)
# Introduce missing values for specified columns
new_df.loc[np.random.choice(new_df.index, size=num_missing), 'membership_level'] = np.nan
```
